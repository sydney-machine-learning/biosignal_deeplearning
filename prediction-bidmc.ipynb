```python
for time_steps in [1,3,5,10,30]:
    print("--------------{}--------------".format(time_steps))
    Xtrain = []
    Ytrain = []
    Test_score = []
    for file in all_csv_files:
        try:
            df = pd.read_csv(file)
            X1, X2, X3, X4 = df[' PLETH'].values, df[' V'].values, df[' AVR'].values, df[' II'].values
            X = np.concatenate([X1.reshape(len(X1),1),X2.reshape(len(X1),1),X3.reshape(len(X1),1),X4.reshape(len(X1),1)], axis=1)
            Y = df[' RESP'].values

            for i in range(X.shape[0]-time_steps+1):
                X_ = X[i:i+time_steps,:]
                Y_ = Y[i+time_steps-1]
                Xtrain.append(X_)
                Ytrain.append(Y_)
        except:
            continue

    for m in range(10):
        X = np.array(Xtrain)
        Y = np.array(Ytrain)

        Xtrain, Xval, Ytrain, Yval = train_test_split(X,Y,test_size=0.2, random_state=42+10*m)

        model = Sequential()
        model.add(LSTM(50, input_shape=(time_steps,4)))
        model.add(Dropout(0.5))
        model.add(Dense(100, activation='swish'))
        # model.add(Dense(50, activation='swish'))
        model.add(Dense(1, activation='linear'))
        model.compile(loss='mean_squared_error', optimizer='adam')
        model.fit(Xtrain, Ytrain,epochs=500, batch_size=10000, verbose=0)
    
        Ypred = model.predict(Xval)
        Ypred = Ypred.reshape((Ypred.shape[0],))
        Test_score.append(np.mean((Ypred-Yval)**2))
        print("Test Error: {}".format(np.mean((Ypred-Yval)**2)))
    print("Best score: {}".format(min(Test_score)))
    print("Average score: {}".format(sum(Test_score)/len(Test_score)))
    print("Variance: {}".format(np.var(Test_score)))
```
